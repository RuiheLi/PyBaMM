```python
# 转换为循环次数 vs SOH 
usage_year = [0, 730]
charge_throughput = [0, 4562.5]
usage_time = [100, 25]

fig, axs = plt.subplots(1,2, figsize=(12,4),tight_layout=True)
axs[0].plot(usage_year, usage_time, linewidth=2,  linestyle='--',marker = 'o',
    markerfacecolor='w',markersize = 8 )
axs[0].set_ylabel("SOH / %", fontproperties=font_prop)
axs[0].set_xlabel("循环次数", fontproperties=font_prop)

axs[1].plot(charge_throughput, usage_time, linewidth=2,  linestyle='--',marker = 'o',
    markerfacecolor='w',markersize = 8 )
axs[1].set_ylabel("SOH / %", fontproperties=font_prop)
axs[1].set_xlabel("电量吞吐量 / Ah", fontproperties=font_prop)

axs[0].text(
    -0.15, 1.04,  "(a)", 
    transform=axs[0].transAxes, 
    fontsize=fs,  
    va='top', ha='right')
axs[1].text(
    -0.15, 1.04,  "(b)", 
    transform=axs[1].transAxes, 
    fontsize=fs,  
    va='top', ha='right')


fig.savefig(Save_fig_path + "/simplest 转换为循环次数 vs SOH .png", dpi=600)
```

```python
# from ref: Ecker, M., Gerschler, J. B., Vogel, J., Käbitz, S., Hust, F., Dechent, P., & Sauer, D. U. (2012). Development of a lifetime prediction model for lithium-ion batteries based on extended accelerated aging test data. Journal of Power Sources, 215, 248-257.
# T range: 25, 35, 50, 65
# V: 3.05, 3.51, 3.92, 4.10
# t: 0~60 weeks 
import numpy as np
def fun_empirical_SOH(T, V, t):
    # input parameters:
    c_a = -0.0064
    c_V = 1.1484
    c_T = 1.5479
    T_0 = 25
    delta_T = 10
    V_0 = 3.5
    delta_V = 0.1
    SOH = (
        1 + 
        c_a * (c_T ** ((T - T_0) / delta_T)) * 
        (c_V ** ((V - V_0) / delta_V)) 
        * np.sqrt(t)  )
    return SOH
t = np.arange(0, 60, 1)
V_list = [3.0, 3.5, 4.0]
T_list = [25, 45, 65]
fig, axs = plt.subplots(1,2, figsize=(12,4),tight_layout=True)
# 固定45°C, 变电压
for i in range(3):
    axs[0].plot(t, fun_empirical_SOH(45, V_list[i], t)*100, 
                linewidth=2, label = f"{V_list[i]} V" )

# 固定3.5V, 变温度
for i in range(3):
    axs[1].plot(t, fun_empirical_SOH(T_list[i], 3.5, t)*100, 
                linewidth=2, label = f"{T_list[i]} °C" )
subplot_labels = ["(a)","(b)"]
titles = ["固定45°C, 变电压", "固定3.5V, 变温度"]
for j in range(2):
    axs[j].set_ylim(70, 105)
    axs[j].legend(prop={'size':fs-1},
              loc='best',frameon=False)
    axs[j].set_ylabel("SOH / %", fontproperties=font_prop)
    axs[j].set_xlabel("搁置时间 / 周", fontproperties=font_prop)
    axs[j].text(
        -0.10, 1.1, 
        subplot_labels[j], transform=axs[j].transAxes, 
        fontsize=fs,  # fontweight='bold', 
        va='top', ha='right')
    axs[j].set_title(titles[j],fontproperties=font_prop)

fig.savefig(Save_fig_path + "/经验公式 - 考虑时间温度和电压 .png", dpi=600)
```
